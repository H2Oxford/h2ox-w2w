```python
reservoirs = ['harangi','kabini','krishnaraja_sagar','bhadra','lower_bhawani','hemavathy']
root = os.path.abspath(os.path.join(os.getcwd(),'..'))
```

```python
params = pickle.load(open(os.path.join(root,'wave2web_data','3zscore_refresh_params.pkl'),'rb'))
```

```python
param_df = pd.DataFrame(params['harangi'])
```

```python
for reservoir in reservoirs:
    param_df = pd.DataFrame(params[reservoir])
    df = pd.read_csv(os.path.join(root,'wave2web_data','refresh_09-27',f'{reservoir}_3zscore.csv')).set_index('Unnamed: 0').drop(columns=['yrmnth','set'])
    df = df*param_df['std']*3+param_df['mean']
    df.to_csv(os.path.join(root,'wave2web_data','refresh_09-27',f'{reservoir}_reproduced.csv'))
```
